Dataset Description:

A total of 688 undergraduate students at Notre Dame University (NDU), Lebanon were selected by simple random sampling method and invited via email to participate in the  'Depression, anxiety, and smartphone
addiction in university students- A cross
sectional study'. They filled various questionnaries which helped make the dataset with the following columns:

1. **ID**: Participant ID for each individual in the dataset.

2. **AGE**: Age of the participant.

3. **Gender**: Gender of the participant (male=0 or female=1).

4. **WrkHrs_Wk**: Weekly work hours of the participant. (0 hours=0, 1-10 hours=1, 11-15 hours=2, >15 hours=3)

5. **Prsnlty_type**: Personality type or traits of the participant. Type A(aggressive, competitive, angry, cynical, mistrustful)=0 and Type B(easy
going, laid-back/ more relaxed, patient)=1

6. **Smoking**: Information about the participant's smoking habits. (No=0, Yes=1)

7. **Alcohol_drnk**: Information about the participant's alcohol consumption. (No=0, Yes=1)

8. **AgeStrt_useSmrtPhne**: Age at which the participant started using a smartphone.

9. **HrsSmrtPhnUse_Wkday**: Average hours of smartphone use on weekdays. (<5 hours=0, >5 hours=1)

10. **Compulsive_Behavior**: Indicates compulsive behavior related to smartphone use.

11. **Functional_Impairment**: Indicates functional impairment due to smartphone use.

12. **Withdrawal**: Reflects withdrawal symptoms when not using a smartphone.

13. **Tolerance**: Indicates tolerance to smartphone use.

14. **Depression_score**: A score related to depression.

15. **Anxiety_score**: A score related to anxiety.

16. **TotAddiction_Score**: Total addiction score, calculated from the previous columns.

Link to the study: https://figshare.com/articles/dataset/Depression_anxiety_and_smartphone_addiction_in_university_students-_A_cross_sectional_study/5278606
